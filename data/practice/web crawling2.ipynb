```python
import time
import random


top_movies=[]
for i in range(35):
    response=requests.get(f'https://pedia.watcha.com/api/staffmades/278/contents?page={i}&size=9',cookies=cookies, headers=headers)
    response=response.json()['result']['result']
    top_movies.append(response)
    time.sleep(random.randrange(1,5))

```

```python
new_list=[]

for i in range(35):
    new_dict_list={'code':top_movies[i][0]['code'],
    'title':top_movies[i][0]['title'],
    'year':top_movies[i][0]['year'],
    'ratings_avg':top_movies[i][0]['ratings_avg'],
    'nations':top_movies[i][0]['nations'][0]['name'],
    'genres':top_movies[i][0]['genres']}
    new_list.append(new_dict_list)


new_list
```

```python
top_movies_df=pd.DataFrame(new_list)
top_movies_df
# top_movies_df.to_csv('top_movies.csv')
```

```python
def get_plot(code):
    response=requests.get(f'https://pedia.watcha.com/api/contents/{code}', cookies=cookies, headers=headers)
    description=response.json()['result']['description']
    return description

top_movies_df['plot']=top_movies_df['code'].progress_apply(get_plot)
```
